```python
import pandas as pd

data = {"x1": [0, 2, 0, 0, -1, 1],
       "x2": [3, 0, 1, 1, 0, 1],
       "x3": [0, 0, 3, 2, 1, 1],
       "Y": ["Red", "Red", "Red", "Green", "Green", "Red"]}

df = pd.DataFrame(data)
df
```

```python
test = [0, 0, 0]
```

```python
import numpy as np

def compute_distance(df: pd.DataFrame, entry: list[int]):
    df["distance"] = (df["x1"] - entry[0])**2 + (df["x2"] - entry[1])**2 + (df["x3"] - entry[2])**2

    return df

def compute_distance(df: pd.DataFrame, entry: list[int]):
    arr = df[['x1', 'x2', 'x3']].to_numpy()
    entry_arr = np.array(entry)
    diff = arr - entry_arr
    df["distance"] = np.sqrt(np.sum(diff ** 2, axis=1))
    return df
```

```python
df2 = compute_distance(df, test)
df2
```

```python
k = 1

df.sort_values("distance").iloc[:k].Y.value_counts().idxmax()
```
